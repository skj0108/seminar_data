```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False 
```

# 범죄발생횟수

```python
data_2017 = pd.read_csv('범죄발생_월_2017.csv')
data_2018 = pd.read_csv('범죄발생_월_2018.csv')
data_2019 = pd.read_csv('범죄발생_월_2019.csv')
```

```python
data_2017.head()
```

```python
data_2017.info()
```

## 다양한 범죄분류 존재 -> 절도, 폭행, 성폭력 살펴보기

```python
data_2017[data_2017['범죄분류'] == '절도'] 
```

```python
data_2017[data_2017['범죄분류'] == '폭행'] 
```

```python
data_2017[data_2017['범죄분류'] == '성폭력'] 
```

```python
data_2017_2 = data_2017.iloc[[0,9,10], :]
data_2017_2
```

```python
data_2018_2 = data_2018.iloc[[0,9,10], :]
```

```python
data_2019_2 = data_2019.iloc[[0,9,10], :]
```

### index, column 위치 바꾸기

```python
data_2017_2 = data_2017_2.set_index('범죄분류')
data_2018_2 = data_2018_2.set_index('범죄분류')
data_2019_2 = data_2019_2.set_index('범죄분류')
```

```python
data_2017_2 = data_2017_2.T
data_2018_2 = data_2018_2.T
data_2019_2 = data_2019_2.T
```

```python
data_2017_2
```

## 월별 절도 발생횟수 시각화

```python
fig = plt.figure(figsize=(10,15))
fig.add_subplot(311)
plt.bar(data_2017_2.index, data_2017_2['절도'])
plt.ylim([10000,20000])
plt.title('2017년 절도 발생 횟수')
fig.add_subplot(312)
plt.bar(data_2018_2.index, data_2018_2['절도'])
plt.ylim([10000,20000])
plt.title('2018년 절도 발생 횟수')
fig.add_subplot(313)
plt.bar(data_2019_2.index, data_2019_2['절도'])
plt.ylim([10000,20000])
plt.title('2019년 절도 발생 횟수')
plt.show()
```

## 월별 성폭력 발생횟수 시각화

```python
fig = plt.figure(figsize=(10,15))
fig.add_subplot(311)
plt.bar(data_2017_2.index, data_2017_2['성폭력'])
plt.title('2017년 성폭력 발생 횟수')
fig.add_subplot(312)
plt.bar(data_2018_2.index, data_2018_2['성폭력'])
plt.title('2018년 성폭력 발생 횟수')
fig.add_subplot(313)
plt.bar(data_2019_2.index, data_2019_2['성폭력'])
plt.title('2019년 성폭력 발생횟수')
plt.show()
```

## 월별 폭행 발생횟수 시각화

```python
fig = plt.figure(figsize=(10,15))
fig.add_subplot(311)
plt.bar(data_2017_2.index, data_2017_2['폭행'])
plt.ylim([10000,17000])
plt.title('2017년 폭행 발생 횟수')
fig.add_subplot(312)
plt.bar(data_2018_2.index, data_2018_2['폭행'])
plt.ylim([10000,17000])
plt.title('2018년 폭행 발생 횟수')
fig.add_subplot(313)
plt.bar(data_2019_2.index, data_2019_2['폭행'])
plt.ylim([10000,17000])
plt.title('2019년 폭행 발생횟수')
plt.show()
```

## 3년 data평균

```python
aver_1 = (data_2017_2['절도'] + data_2018_2['절도'] + data_2019_2['절도']) / 3
aver_1
```

```python
aver_2 = (data_2017_2['성폭력'] + data_2018_2['성폭력'] + data_2019_2['성폭력']) / 3
aver_3 = (data_2017_2['폭행'] + data_2018_2['폭행'] + data_2019_2['폭행']) / 3
```

```python
fig = plt.figure(figsize=(15,10))
fig.add_subplot(221)
plt.bar(aver_1.index, aver_1)
plt.ylim([10000,20000])
plt.title('절도 발생 횟수')
plt.grid(True, axis='y')
fig.add_subplot(222)
plt.bar(aver_2.index, aver_2)
plt.title('성폭력 발생 횟수')
plt.ylim([0,5000])
plt.grid(True, axis='y')
fig.add_subplot(223)
plt.bar(aver_3.index, aver_3)
plt.ylim([10000,17000])
plt.title('폭행 발생 횟수')
plt.grid(True, axis='y')
plt.show()
```

# 범행동기

```python
motive_2017 = pd.read_csv('범죄자_범행동기_2017.csv')
motive_2018 = pd.read_csv('범죄자_범행동기_2018.csv')
motive_2019 = pd.read_csv('범죄자_범행동기_2019.csv')
motive_2019.head(11)
```

## 절도, 성폭력, 폭행 추출

```python
motive_2017_2 = motive_2017.iloc[[0,9,10],:]
motive_2018_2 = motive_2018.iloc[[0,9,10],:]
motive_2019_2 = motive_2019.iloc[[0,9,10],:]
```

```python
motive_2017_2 = motive_2017_2.set_index('범죄분류')
motive_2018_2 = motive_2018_2.set_index('범죄분류')
motive_2019_2 = motive_2019_2.set_index('범죄분류')

motive_2017_2 = motive_2017_2.T
motive_2018_2 = motive_2018_2.T
motive_2019_2 = motive_2019_2.T
```

### 절도 범행동기

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(131)
plt.bar(motive_2017_2.index, motive_2017_2['절도'])
plt.xticks(rotation=90)
plt.title('2017년 절도 범행동기')
fig.add_subplot(132)
plt.bar(motive_2018_2.index, motive_2018_2['절도'])
plt.xticks(rotation=90)
plt.title('2018년 절도 범행동기')
fig.add_subplot(133)
plt.bar(motive_2019_2.index, motive_2019_2['절도'])
plt.xticks(rotation=90)
plt.title('2019년 절도 범행동기')
plt.show()
```

## 성폭력 범행동기

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(131)
plt.bar(motive_2017_2.index, motive_2017_2['성폭력'])
plt.xticks(rotation=90)
plt.title('2017년 성폭력 범행동기')
fig.add_subplot(132)
plt.bar(motive_2018_2.index, motive_2018_2['성폭력'])
plt.xticks(rotation=90)
plt.title('2018년 성폭력 범행동기')
fig.add_subplot(133)
plt.bar(motive_2019_2.index, motive_2019_2['성폭력'])
plt.xticks(rotation=90)
plt.title('2019년 성폭력 범행동기')
plt.show()
```

## 폭행 범행동기

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(131)
plt.bar(motive_2017_2.index, motive_2017_2['폭행'])
plt.xticks(rotation=90)
plt.title('2017년 폭행 범행동기')
fig.add_subplot(132)
plt.bar(motive_2018_2.index, motive_2018_2['폭행'])
plt.xticks(rotation=90)
plt.title('2018년 폭행 범행동기')
fig.add_subplot(133)
plt.bar(motive_2019_2.index, motive_2019_2['폭행'])
plt.xticks(rotation=90)
plt.title('2019년 폭행 범행동기')
plt.show()
```

절도, 성폭력은 우발적 범죄가 가장 많고, 폭행 또한 우발적 범죄가 2번째로 높은 순에 속합니다.
